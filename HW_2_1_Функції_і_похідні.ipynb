1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np

# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 10x - 5:", derivative_f_lin)
```

```python

import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10*x - 5
df = 10 * np.ones(len(x)) # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green') 

# Додавання міток та заголовка
plt.title("Графік функції (f(x) = 10*x - 5) та її похідної (f'(x) = 10)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

### Висновок: Якщо похідна показує зміну в актуальній точці часу і дорівнює 10, то у кожній точці навчання збільшення оцінки є на 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 1000)

f = np.sin(x)
df = np.cos(x)

plt.figure(figsize=(10,5))

plt.plot(x, f, label='f(x)=sin(x)')
plt.plot(x, df, label="f'(x)=cos(x)")

plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 1000) 
f = np.log(x)
df = 1/x

plt.figure(figsize=(10,5))

plt.plot(x, f, label='f(x)=ln(x)')
plt.plot(x, df, label="f'(x)=1/x")

plt.legend()
plt.grid()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

# Лінійна функція
f_lin = x**2 - 5*x + 1
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 'x**2 - 5*x + 1': ", derivative_f_lin)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = x**2 - 5*x + 1
df = 2*x - 5

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x**2 - 5*x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2*x - 5", color='green') 

# Додавання міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```
